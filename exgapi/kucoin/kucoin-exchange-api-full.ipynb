```python
import sys
sys.path.append("../../lib")
from KucoinAPI import *
```

# Global variables

```python
with open("../../../keys/kc-key") as f:
    api_key, api_secret, api_passphrase = f.read().split()
```

# Load kucoin API

```python
kapi = KucoinAPI(api_key, api_secret, api_passphrase)
```

# Public api endpoints

```python
# Get all symbols
symbols = kapi.getTradeSymbols()
```

```python
symbols
```

```python
# get orderbook
kapi.getOrderbook("BNB-USDT", 20)
```

```python
# get trade history
kapi.getTradeHistory("BTC-USDT")
```

```python
# get klines
etime = int(time.time())
stime = etime - 3600 * 24
kapi.getKlines("1hour", "BTC-USDT", stime=stime, etime=etime)
```

# Private api endpoints

```python
# get account info
kapi.getAccountInfo(accountType="main")
```

```python
# transfer asset between accounts
kapi.internalTransfer(currency="BNB", from_="main", to_="trade", amount=0.05)
```

```python
# get account info
kapi.getAccountInfo(accountType=None)
```

```python
# place order
orderid = kapi.placeOrder(side="sell", symbol="BNB-USDT", order_type="limit", price=500, size=0.05)
```

```python
orderid
```

```python
# get order
kapi.getOrder(orderid)
```

```python
# cancel order
kapi.cancelOrder(orderid)
```

```python
# get order
kapi.getOrder(orderid)
```

```python
# place stop order
stopid = kapi.placeStopOrder(side="sell", symbol="BNB-USDT", stop_price=400, price=399, size=0.05, order_type="limit", stop_type="loss")
```

```python
stopid
```

```python
# get order
kapi.getOrder(stopid)
```

```python
# cancel order
kapi.cancelOrder(stopid)
```

```python
# get order
kapi.getOrder(stopid)
```
